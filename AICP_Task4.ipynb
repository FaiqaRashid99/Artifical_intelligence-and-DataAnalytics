# Q.1:Import libraries (Numpy, pandas, matplotlib, plotly and seaborn) and then read csv file.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
# Read CSV file
df = pd.read_csv('/content/drive/MyDrive/menu.csv')

# Display first few rows of the dataset
df.head()
```

# Q.2: Check **statistical facts** by checking all columns. Then calculate the **maximum value** of the following attributes
 ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']

```python
# Display statistical facts
print(df.describe())

# Calculate maximum values for specified attributes
max_values = df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']].max()

print("Maximum values:")
print(max_values)
```

# Q.3: Check to see if infact there is any **correlation** between Calories and other independent variables by plotting a correlation matrix next.

```python
plt.figure(figsize=(12, 8))
df_f=df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']]
corr = df_f.corr()
sns.heatmap(corr, annot=True, cmap='rocket', fmt='.2f')
plt.title('Correlation Matrix')
plt.xticks(rotation=60)
plt.show()
```

# Q.4: Draw **boxplot for Calories vs Category** to spot outliers and max calories category


```python
plt.figure(figsize=(10, 8), facecolor='white')
sns.boxplot(x='Calories', y='Category', data=df, orient='h',hue='Category')
plt.title('Boxplot of Calories by Category')
plt.xlabel('Calories')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

# Q.5: Figure out which exact item contains a **high quantity** for:
['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium
(% Daily Value)', 'Iron (% Daily Value)']. Display summary

```python
# Find the item with maximum values for specified attributes
max_calories_item = df.loc[df['Calories'].idxmax()]
max_nutrients_item = df.loc[df[['Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']].idxmax()]

print("Item with Maximum Calories:")
print(max_calories_item[['Item', 'Calories']])

print("\nItem with Maximum Nutrient Values:")
print(max_nutrients_item[['Item', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']])
```

# Q.6: Draw **Stripplot** for each category against the following attributes
['Calories', 'Total Fat',
'Carbohydrates', 'Dietary
Fiber', 'Sugars', 'Protein',
'Vitamin A (% Daily Value)',
'Vitamin C (% Daily Value)',
'Calcium (% Daily Value)', 'Iron
(% Daily Value)

```python
# Select attributes for strip plots
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']

# Create strip plots for each category and each attribute
for attribute in attributes:
    plt.figure(figsize=(10, 5))
    sns.stripplot(x='Category', y=attribute, hue='Category', data=df, jitter=True, alpha=0.7)
    plt.title(f'{attribute} by Category (Stripplot)')
    plt.xticks(rotation=60)
    plt.show()
```

# Q.7: Draw a **horizontal bar graph** for items in each category against the calories.
Also, write your observation.


```python
# Get unique categories
categories = df['Category'].unique()

# Loop through each category and draw a horizontal bar graph
for category in categories:
    category_data = df[df['Category'] == category]

    plt.figure(figsize=(8, 8))
    plt.barh(category_data['Item'], category_data['Calories'], color='skyblue')
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.title(f'Calories for Items in {category} Category')
    plt.show()
```
