# Q.1: Writer pandas series code to get following output without using dictionary:
* a   1
* x   4
* c   9
* 2   6
* e   7

```python
import pandas as pd

s = pd.Series([1, 4, 9, 6, 7])
s.index = ['a', 'x', 'c', 2, 'e']

print(s)
#Pandas displays both the index and the values in a tabular format.
```

# Q3. Write pandas dataframe code to get following output using python dictionary
* day  temperature  windspeed  event

*   0   1/1/2017   32   6   Rain
*   1   1/2/2017   35   7   Sunny
*   2   1/3/2017   28   2   Snow
*   3   1/4/2017   24   7   Snow
*   4   1/5/2017   32   4   Rain
*   5   1/6/2017   31   2   Sunny

```python
# Define a dictionary named "data"
# where 'keys' are column names and 'values' are lists of data for each column.
data = {'day': ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017', '1/5/2017', '1/6/2017'],
        'temperature': [32, 35, 28, 24, 32, 31],
        'windspeed': [6, 7, 2, 7, 4, 2],
        'event': ['Rain', 'Sunny', 'Snow', 'Snow', 'Rain', 'Sunny']}

#convert dictionary to dataframe:
df = pd.DataFrame(data)
df
```

# Q.4: In extension to above question, you are required to replace index by ['a','b','c','d','e','f']

```python
data = {'day': ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017', '1/5/2017', '1/6/2017'],
        'temperature': [32, 35, 28, 24, 32, 31],
        'windspeed': [6, 7, 2, 7, 4, 2],
        'event': ['Rain', 'Sunny', 'Snow', 'Snow', 'Rain', 'Sunny']}

# Creating DataFrame with custom index
df = pd.DataFrame(data, index=['a', 'b', 'c', 'd', 'e', 'f'])

df
```

# Q.5: In extension to above Q.3, calculate mean, miximum and minimum for label “temperature”

```python
# Calculate mean, maximum, and minimum for the 'temperature' column
mean_temperature = df['temperature'].mean()
max_temperature = df['temperature'].max()
min_temperature = df['temperature'].min()

# Print the results
print(f"Mean Temperature: {mean_temperature}")
print(f"Maximum Temperature: {max_temperature}")
print(f"Minimum Temperature: {min_temperature}")
```

 Q.8: Create the following dataframe as AICP_DF then implement different operations as described below:

---


* select 'Name', 'Qualification' coloumns and save to df1

---

### dataframe
* Name   Age   Address   Qualification
* Sonia   27   Lahore   Msc
* Bilal   24   Karachi   MA
* Hifza   22   Sialkot   MCA
* Kabir   32   Peshawar   Phd
* Jazim   23   lhr   bsc


```python
# Creating AICP_DF DataFrame (using dictionary "key:value")
data = {'Name': ['Sonia', 'Bilal', 'Hifza', 'Kabir', 'Jazim'],
        'Age': [27, 24, 22, 32, 23],
        'Address': ['Lahore', 'Karachi', 'Sialkot', 'Peshawar', 'Lahore'],
        'Qualification': ['Msc', 'MA', 'MCA', 'Phd', 'Bsc']}
AICP_DF = pd.DataFrame(data)
AICP_DF
```

```python
# Selecting 'Name' and 'Qualification' columns and saving to df1
df1 = AICP_DF[['Name', 'Qualification']]
df1
```

```python
# Adding a new column 'Height' to AICP_DF, values: [5.1, 6.2, 5.1, 5.2,5.1]
AICP_DF['Height'] = [5.1, 6.2, 5.1, 5.2, 5.1]
AICP_DF
```

```python
# Setting 'Name' as the index column
AICP_DF.set_index('Name', inplace=True)
AICP_DF
```

```python
# Retrieving row with index 'Hifza'
row_hifza = AICP_DF.loc['Hifza']
print("\nRow with index 'Hifza':\n", row_hifza)
```

```python
# Retrieving row with index 3
row_index_3 = AICP_DF.iloc[3]  # Note: Python uses 0-based indexing
row_index_3
```

```python
# Dropping row with index 'Bilal'
AICP_DF.drop(index='Bilal', inplace=True)
AICP_DF
```
