**Q.7: Have a look at the distribution of reach from different sources**

```python
source_columns = ['From Home', 'From Hashtags', 'From Explore', 'From Other']

# Replace NaN values with 0 in source columns
df[source_columns] = df[source_columns].fillna(0)

# Calculate the total reach for each post
df['Total Reach'] = df[source_columns].sum(axis=1)

# Create a new DataFrame with columns 'Source' and 'Total Reach'
source_df = pd.DataFrame({
    'Source': source_columns,
    'Total Reach': df[source_columns].sum()
})

fig = px.pie(source_df, names='Source', values='Total Reach', title='Distribution of Reach from Different Sources')
fig.show()
```

**Q.8: Have a look at the distribution of engagement sources**

```python
engagement_columns = ['Likes', 'Saves', 'Shares', 'Comments']

# Calculate total engagement for each post
df['Total Engagement'] = df[engagement_columns].sum(axis=1)

# Create a new DataFrame with columns 'Source' and 'Total Engagement'
engagement_df = pd.DataFrame({
    'Source': engagement_columns,
    'Total Engagement': df[engagement_columns].sum()
})
# Define a custom color scale for the pie chart
custom_colors = px.colors.qualitative.Set3

fig = px.pie(engagement_df, names='Source', values='Total Engagement', title='Distribution of Engagement Sources',color_discrete_sequence=custom_colors)
fig.show()
```

 **Q.9: Have a look at the relationship between the number of profile visits and follows**

```python
import numpy as np
# Scatter plot
fig = px.scatter(df, x='Profile Visits', y='Follows', title='Relationship between Profile Visits and Follows')

# Calculate the best-fit line (linear regression)
m, b = np.polyfit(df['Profile Visits'], df['Follows'], 1)
best_fit_line = m * df['Profile Visits'] + b

# Overlay the best-fit line
fig.add_trace(go.Scatter(x=df['Profile Visits'], y=best_fit_line, mode='lines', name='Best-Fit Line'))

# Show figure
fig.show()
```

**Q.10: Have a look at the type of hashtags used in the posts using a wordcloud**

**Q.12: Havea look at the distribution of hashtags to see which hashtag is used the most in all the posts**

```python
# Get all the hashtags from the DataFrame
hashtags = df['Hashtags'].dropna()

# Create a dictionary to store the frequency of each hashtag
hashtag_counts = {}
for hashtag_list in hashtags:
    for hashtag in hashtag_list.split():
        if hashtag not in hashtag_counts:
            hashtag_counts[hashtag] = 0
        hashtag_counts[hashtag] += 1

# Sort the dictionary by the frequency of each hashtag
sorted_hashtags = sorted(hashtag_counts.items(), key=lambda x: x[1], reverse=True)

# Create a bar chart to visualize the frequency of each hashtag
fig = px.bar(sorted_hashtags, x=0, y=1, title='Distribution of Hashtags')
fig.update_xaxes(tickangle=-45)
fig.show()
```

 **Q.13: Have a look at the distribution of likes and impressions received from the presence of each hashtag on the post**

```python
# Create two new DataFrames for likes and impressions
likes_df = df.groupby('Hashtags')['Likes'].sum().reset_index()
impressions_df = df.groupby('Hashtags')['Impressions'].sum().reset_index()

# Create a bar chart for the distribution of likes
fig = px.bar(likes_df, x='Hashtags', y='Likes', title='Distribution of Likes by Hashtag')
# fig.update_xaxes(tickangle=-45)
fig.show()

# Create a bar chart for the distribution of impressions
fig = px.bar(impressions_df, x='Hashtags', y='Impressions', title='Distribution of Impressions by Hashtag')
# fig.update_xaxes(tickangle=-45)
fig.show()
```
